```python
#kevin
# dataset = {'307': 1, '302': 2, '427':3,'34':4,'3':5,'4':8,'27':7,'87':8,'1':10,'290':10,'2':5,'97':12,'13':13,'142':14,'14221':15,'142':16}
# dataset = {'65': 1, '57': 2,'35': 3,'226': 4,'2': 5,'3':5,'19':7,'4':8,'26':9,'1':10}
#aman
# dataset = {'14': 1, '16': 2, '5': 3, '4': 8, '309': 5, '371': 6, '2': 5, '1': 10, '22': 9, '15': 10, '37': 11, '38': 12, '47': 13, '51': 14}
# dataset = {'65': 2, '80': 4, '81':6,'4':8,'3':5,'6':5,'63':12,'2':5,'7':6,'62':18,'1659':20,'165':21,'710':22,'91':23}
#shubham
# dataset = {'2': 5,'6': 5,'21': 3,'13': 4,'15': 5,'16': 6,'24': 7,'25': 8,'120': 9,'228': 10, '215': 11,'235': 12,'376': 13,'30': 14,'375': 15}
# dataset = {'2': 5, '3': 5,'4': 8,'3415': 4,'6': 5,'7': 6,'354': 7}

dataset ={'307': 1, '302': 2, '427':3,'34':4,'3':5,'4':8,'27':7,'87':8,'1':10,'290':10,'2':5,'97':12,'13':13,'142':14,'14221':15,'142':16,'65': 1, '57': 2,'35': 3,
          '226': 4,'2': 5,'3':5,'19':7,'4':8,'26':9,'1':10,'14': 1, '16': 2, '5': 3, '4': 8, '309': 5, '371': 6, '2': 5, '1': 10, '22': 9, '15': 10, '37': 11, '38': 12,
          '47': 13, '51': 14,'65': 2, '80': 4, '81':6,'4':8,'3':5,'6':5,'63':12,'2':5,'7':6,'62':18,'1659':20,'165':21,'710':22,'91':23,'2': 5,'6': 5,'21': 3,'13': 4,
          '15': 5,'16': 6,'24': 7,'25': 8,'120': 9,'228': 10, '215': 11,'235': 12,'376': 13,'30': 14,'375': 15,'2': 5, '3': 5,'4': 8,'3415': 4,'6': 5, '7':6, '354':7 }
count=0
s1=[]
a=len(results[0]["annotations"])
for i in range(a):
  if "entities" in results[0]["annotations"][i]:
    length=len(results[0]["annotations"][i]["entities"])
    # print(length)
  
    if(length>0):
      for j in range(length): 
        # print(results[0]["annotations"][i]["entities"][j]["labels"],"\n")
        if(results[0]["annotations"][i]["entities"][j]["labels"][0]=="PROVISION"):
          count=count+1

          s1.append(results[0]["annotations"][i]["entities"][j]["text"])
          # print(results[0]["annotations"][i]["entities"][j]["text"])

import re
pattern = r"\d+"
# Extract numbers from each string using regular expressions
numbers = []
for string in s1:
    matches = re.findall(pattern, string)
    numbers.extend(matches)

# Print the resulting list of numbers
print(numbers)

```

```python
from collections import defaultdict
# Create a defaultdict with int as the default factory
word_counts = defaultdict(int)

# Loop through each word in the list and increment its count in the dictionary
for word in numbers:
    word_counts[word] += 1

# Print the resulting dictionary
#print(word_counts)
sorted_word_counts = sorted(word_counts.items(), key=lambda x:x[1])
length=len(sorted_word_counts)
print(sorted_word_counts)


```

```python
#printing result after multiplying the values with the dataset
result = []
for key, value in sorted_word_counts:
    if key in dataset:
        result.append((key, value * dataset[key]))

print(result)
```

```python
max_tuple = max(result, key=lambda x: x[1])
res = max_tuple[0]

print('Article '+res)
```
